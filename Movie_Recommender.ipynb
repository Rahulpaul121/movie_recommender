```python
import numpy as np
import pandas as pd
import ast
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.shape
```

```python
credits.shape
```

```python
movies.info()
```

```python
# merging the 2 dataset together based of title
movies=movies.merge(credits,on="title")
```

```python
movies.info()
```

```python
#Keeping relevent columns 
#generes
#id
#title
#keywords
#overview
#cast
#crew
movies=movies[['genres','id','title','keywords','overview','cast','crew']]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3 :
            L.append(i['name'])
            count+=1
        else:
            break
    return L
```

```python
movies["cast"]=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fectch_dir(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director' :
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fectch_dir)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tag']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['id','title','tag']]
```

```python
new_df.head()
```

```python
new_df['tag']=new_df['tag'].apply(lambda x:" ".join(x))
```

```python
new_df['tag']=new_df['tag'].apply(lambda x: x.lower())
```

```python
new_df['tag']
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tag']=new_df['tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tag']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommender (movie):
    movie_index= new_df[new_df['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommender('Batman Begins')
```
